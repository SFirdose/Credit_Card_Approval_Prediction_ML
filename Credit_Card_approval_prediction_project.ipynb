## Outlier Detection
### Using the IQR Method
1. When the data was normally distributed we use Z-score method to detect the outliers.
2. When the data was skewed we use IOR Method.
3. Now for Age,Annual_income, Family_members and Children columns we use IQR method to fix the outliers because all these columns are skewed.
4. Removal of outliers are must because it effects the model and won't give the correct accuracy because the outliers effects the data.

```python
cards['Annual_income'].describe()
```

```python
sns.boxplot(cards['Annual_income'])
```

```python
# Finding the IQR
percentile25 = cards['Annual_income'].quantile(0.25)
percentile75 = cards['Annual_income'].quantile(0.75)
```

```python
iqr = percentile75 - percentile25
```

```python
iqr 
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

```python
cards[cards['Annual_income'] > upper_limit]
```

```python
cards[cards['Annual_income'] > lower_limit]
```

```python
new_df_card = cards.copy()

new_df_card['Annual_income'] = np.where(
    new_df_card['Annual_income'] > upper_limit,
    upper_limit,
    np.where(
        new_df_card['Annual_income'] < lower_limit,
        lower_limit,
        new_df_card['Annual_income']
    )
)
```
